# **DATA CLEANING**

```python
df1[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']] = df1[['Glucose','BloodPressure','SkinThickness','Insulin','BMI']].replace(0,np.NaN)
```

```python
df1.head()
```

```python
df1.isnull().sum()
```

```python
df1
```

```python
import missingno as msno
msno.bar(df1);
```

```python
# The missing values will be filled with the median values of each variable.
def median_target(var):
    temp = df1[df1[var].notnull()]
    temp = temp[[var, 'Outcome']].groupby(['Outcome'])[[var]].median().reset_index()
    return temp
# The values to be given for incomplete observations are given the median value of people who are not sick and the median values of people who are sick.
columns = df1.columns
columns = columns.drop("Outcome")
for i in columns:
    median_target(i)
    df1.loc[(df1['Outcome'] == 0 ) & (df1[i].isnull()), i] = median_target(i)[i][0]
    df1.loc[(df1['Outcome'] == 1 ) & (df1[i].isnull()), i] = median_target(i)[i][1]
```

```python
# In the data set, there were asked whether there were any outlier observations compared to the 25% and 75% quarters.
# It was found to be an outlier observation.
for feature in df1:

    Q1 = df1[feature].quantile(0.25)
    Q3 = df1[feature].quantile(0.75)
    IQR = Q3-Q1       #interquartile range
    lower = Q1- 1.5*IQR
    upper = Q3 + 1.5*IQR

    if df1[(df1[feature] > upper)].any(axis=None):
        print(feature,"yes")
    else:
        print(feature, "no")
```

```python
# The process of visualizing the Insulin variable with boxplot method was done. We find the outlier observations on the chart.
import seaborn as sns
sns.boxplot(x = df1["Insulin"]);
```

```python
#We conduct a stand alone observation review for the Insulin variable
#We suppress contradictory values  #interquatrile range
Q1 = df1.Insulin.quantile(0.25)
Q3 = df1.Insulin.quantile(0.75)
IQR = Q3-Q1
lower = Q1 - 1.5*IQR
upper = Q3 + 1.5*IQR
df1.loc[df1["Insulin"] > upper,"Insulin"] = upper
```
